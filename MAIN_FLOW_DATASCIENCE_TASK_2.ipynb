**DATA CLEANING**

```python
#dropping the duplicates
data=data.drop_duplicates()
data
```

```python
#Detecting Missing Values in a DataFrame
data.isnull()
```

```python
#Counting Missing Values in Each Column of a DataFrame
data.isnull().sum()
```

```python
#Detecting Non-Missing Values in a DataFrame
data.notnull()
```

```python
#Counting Total Missing Values in a DataFrame
data.isnull().sum().sum()
```

```python
#Replacing Missing Values with a Specified Value in a DataFrame
data1=data.fillna(value=0)
data1
```

```python
#Verifying Total Missing Values After Imputation
data1.isnull().sum().sum()
```

```python
#Forward Fill Missing Values in a DataFrame
data1=data.fillna(method='pad')
data1
```

```python
data1=data.fillna(method='ffill')
data1
```

```python
#Backward Fill Missing Values
data1=data.fillna(method='bfill')
data1
```

**PREPROCESSING**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Retrieving Column Names of a DataFrame
data1.columns
```

```python
#Dropping a Column from a DataFrame
data1.drop(['Observation'], axis=1, inplace=True)
data1.columns
```

```python
#Calculating the Interquartile Range (IQR) of a DataFrame.
Q1=data1.quantile(0.25)
Q3=data1.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
#Removing Outliers from a DataFrame Using IQR
data1=data1[~((data1<(Q1-1.5*IQR))|(data1>(Q3+1.5*IQR))).any(axis=1)]
data1
```
